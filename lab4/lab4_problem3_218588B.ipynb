```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
# Get the files that exist in a directory
def get_filelist(exfile_path):
    files = os.listdir(exfile_path)
    for i, file in enumerate(files):
        a, b = file.split(".")
        if b != "dat":
            files.pop(i)
    return files
```

```python
def calc_file_mean(_file_path):
    file_mean = 0
    # import data
    df = pd.read_table(_file_path, header=None)
    # Convert list and Calculate mean
    
    import statistics
    value_list = [float(s) for s in list(df[0])]
    # x1_stdev = statistics.stdev(x1)
    file_mean = statistics.mean(value_list)
    
    return file_mean
```

```python
# import data file name
data_files = get_filelist('./data')

# Separate h.dat file and j.dat file.
h_dat_file_value = {}

j_dat_file_value = {}

# h.dat file is 0 (Y=0) and j.dat file is 1 (Y=1)
for i in data_files:
    _tmp = []
    if i[0].split('.')[0][0] == 'h':
        _tmp.append(calc_file_mean('./data/'+i))
        _tmp.append(0)
        h_dat_file_value[i] = _tmp
    else:
        _tmp.append(calc_file_mean('./data/'+i))
        _tmp.append(1)
        j_dat_file_value[i] = _tmp
```

```python
print('-----------Check the h.dat means-----------')
for key, values in h_dat_file_value.items():
    print("Key：{}, Values：{}".format(key,values))
```

```python
print('-----------Check the j.dat means-----------')
for key, values in j_dat_file_value.items():
    print("Key：{}, Values：{}".format(key,values))
```

```python
def judge_gorup(x_split, x_data):
    _group = []
    for i in x_data:
        if i <= x_split:
            _group.append("LEFT")
        else:
            _group.append("RIGHT")
    return _group
```

```python
def decision_stump(x, y, x_split, weight):
    result = {}
    group = judge_gorup(x_split[0], x)
    result['group'] = group
    
    print('=====Group=====')
    for i in group:
        print(i)

    prediction = []
    print('=====Prediction=====')
    for n in group:
        if n == 'LEFT':
            print(x_split[1])
            prediction.append(x_split[1])
        else:
            print(abs(x_split[1]-1))
            prediction.append(abs(x_split[1]-1))
    result['prediction'] = prediction
    
    error = []

    print('=====Error=====')
    for i in range(len(prediction)):
        print(abs(int(prediction[i]) - int(y[i])))
        error.append(abs(int(prediction[i]) - int(y[i])))
    result['error'] = error
    
    weight_error = []

    print('=====Weight Error=====')
    for i in range(len(error)):
        print(int(error[i])*float(weight[i]))
        weight_error.append(int(error[i])*float(weight[i]))
    result['weight_error'] = weight_error
    
    import math
    print('=====Sum Weights=====')
    sum_w = math.ceil(sum(weight))
    print(sum_w)
    result['sum_w'] = sum_w

    print('=====Sum Errors=====')
    sum_e = (sum(weight_error))
    print(sum_e)
    result['sum_e'] = sum_e

    print('=====Misclassification=====')
    misclassification = sum_e/sum_w
    print(misclassification)
    result['misclassification'] = misclassification
    
    print('=====Stage=====')
    stage = math.log((1-misclassification)/misclassification)
    print(stage)
    result['stage'] = stage
    
    print('=====Accuracy=====')
    print((1-sum(error)/len(error))*100)
    result['Accuracy'] = (1-sum(error)/len(error))*100
    
    return result
```

```python
x1 = []
y1 = []
weight = []

for i in j_dat_file_value.values():
    x1.append(i[0])
    y1.append(i[1])
    weight.append(1/(len(j_dat_file_value)+len(h_dat_file_value)))

for i in h_dat_file_value.values():
    x1.append(i[0])
    y1.append(i[1])
    weight.append(1/(len(j_dat_file_value)+len(h_dat_file_value)))
result1 = decision_stump(x1, y1, [x1[19],y1[19]], weight)
```

```python
def calc_weighted_stage(prediction, stage):
    model = []
    for i in prediction:
        if i == 1:
            model.append(stage*1)
        else:
            model.append(stage*(-1))
    return model
```
